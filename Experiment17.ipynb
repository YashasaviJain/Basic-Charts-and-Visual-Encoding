```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)
print(df)
```

```python
#1.Line chart(Trend)
plt.plot(df['Days'], df['Study_Hours'], marker='o')
plt.xlabel('Days')
plt.ylabel('Study Hours')
plt.title('Study Hours Trend')
plt.show()
```

```python
# Advanced Line Chart
plt.figure(figsize=(7, 4))

plt.plot(df['Days'], df['Study_Hours'], marker='o', label='Study Hours',color='blue')
plt.plot(df['Days'], df['Marks'], marker='s', label='Marks', color='red')
plt.xlabel('Days')
plt.ylabel('Value')
plt.title('Advanced Line Chart')
plt.legend()
```

```python
#Bar chart (Comprison)
plt.bar(df['Days'], df['Marks'], label='Study Hours')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.title('Marks Comparison by Day')
plt.show()
```

```python
#Advanced Bar Graph
bars = plt.bar(df['Days'], df['Marks'],color = 'red')
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, str(yval), ha='center', va='bottom')
plt.title("Shardul Rasal")
plt.xlabel("Days")
plt.ylabel("Marks")
plt.grid(axis = 'y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(len(df['Days'])) #positions fors bars
width=0.35 #width of each bar
plt.bar(x-width/2,df['Study_Hours'],width,label='Study Hours')
plt.bar(x+width/2,df['Marks'],width,label='Marks')
plt.xticks(x,df['Days'])
plt.xlabel('Days')
plt.ylabel('Values')
plt.title('Shardul')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data (more values for better histogram)

marks = [50, 60, 55, 70, 65, 75, 80, 85, 90, 45, 68, 72, 77, 82, 88]

# Calculate mean

mean_value = np.mean(marks)

plt.figure(figsize=(7,4))

plt.hist(marks, bins=6, color='darkblue', edgecolor='black')

# Mean line

plt.axvline(mean_value, color='red', linestyle='--', linewidth=2, label='Mean')

plt.title("Shardul")

plt.xlabel("Marks")

plt.ylabel("Frequency")

plt.legend()

plt.grid(axis='y', linestyle='--', alpha=0.5)
```

```python
#4. Scatter Plot (Relationship)
import matplotlib.pyplot as plt
import pandas as pd

data = {
    'Days': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Study_Hours': [2, 3, 2.5, 4, 3.5],
    'Marks': [50, 60, 55, 70, 65],
    'Attendance': [60, 70, 65, 80, 75],
    'Sleep_Hours': [6, 7, 6, 8, 7],
    'Assignments_Completed': [1, 2, 1, 3, 2]
}
df = pd.DataFrame(data)

plt.scatter(df['Study_Hours'], df['Marks'])
plt.title("Shardul")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt

#Conditional scatter plot

#Example category

df['Result'] = ['Fail', 'Pass', 'Fail', 'Pass', 'Pass']

colors = ['red' if r=='Fail' else 'green' for r in df['Result']]

plt.scatter(df['Study_Hours'], df['Marks'], c=colors)

plt.title("Study Hours vs Marks (Result-wise)")

plt.xlabel("Study Hours")

plt.ylabel("Marks")

plt.show()
```
